```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.read_csv('data/train.csv')
data.head(50)
```

```python
# Just use the text and target column
data = data[['text', 'target']]
```

```python
# Lowercase the text
data['text'] = data['text'].str.lower()
data.head()
```

```python
data.head(50)
```

```python
dataprep1 = data.copy()
```

```python
import re
# Remove urls
dataprep1['text'] = dataprep1['text'].apply(lambda x: re.sub(r'https?://\S+|www\.\S+', '', x))
dataprep1.head()
```

```python
dataprep1.head(50)
```

```python
dataprep2 = dataprep1.copy()
```

```python
# Remove html tags
dataprep2['text'] = dataprep2['text'].apply(lambda x: re.sub(r'<.*?>', '', x))
dataprep2.head()
```

```python
# Remove punctuations
dataprep2['text'] = dataprep2['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
dataprep2.head()
```

```python
dataprep2.head(50)
```

```python
dataprep2.to_csv('data/disaster.csv', index=False)
```
